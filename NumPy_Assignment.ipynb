```python
import numpy as np

# Expense dataset
expenses = np.array([1500, 2000, 1800, 2200, 1700, 2500])

# 1. Average monthly expense
average_expense = np.mean(expenses)
print("Average monthly expense:", average_expense)

# Reflection:
# The average monthly expense is 1950.0, indicating the typical amount spent each month.

# 2. Highest expense month
highest_expense = np.max(expenses)
highest_month = np.argmax(expenses) + 1
print("Highest expense:", highest_expense)
print("Month with highest expense:", highest_month)

# Reflection:
# Month 6 had the highest expense of 2500, making it the most expensive month.

# 3. Lowest expense month
lowest_expense = np.min(expenses)
lowest_month = np.argmin(expenses) + 1
print("Lowest expense:", lowest_expense)
print("Month with lowest expense:", lowest_month)

# Reflection:
# Month 1 had the lowest expense of 1500, making it the least expensive month.

# 4. Total expenses
total_expenses = np.sum(expenses)
print("Total expenses:", total_expenses)

# Reflection:
# The total expenses for the six months amounted to 11700.

# 5. Slice to show only the first 3 months
first_three_months = expenses[:3]
print("First three months' expenses:", first_three_months)

# Reflection:
# The expenses for the first three months were 1500, 2000, and 1800.
```
