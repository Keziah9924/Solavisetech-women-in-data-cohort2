```python
import pandas as pd
import numpy as np

data = {
    'Order_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Customer_Name': ['Amara', 'Bola', None, 'Diana', 'Efua', 'Fatima', 'Grace',
                      None, 'Ifeoma', 'Joyce'],
    'Country': ['Ghana', 'Nigeria', 'Kenya', None, 'Ghana', 'Cameroon', None,
                'Kenya', 'Nigeria', 'Ghana'],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', None, 'Phone', 'Tablet',
                'Laptop', None, 'Phone'],
    'Quantity': [2, 1, None, 3, 2, None, 1, 4, 2, None],
    'Price': [1500, 800, 600, 1500, None, 800, 600, 1500, None, 800],
    'Delivery_Days': [3, None, 7, 5, 3, None, 6, None, 4, 5]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
Explanation

The dataset contains 10 records and 7 columns. Several columns contain missing values, including Customer_Name, Country, Product, Quantity, Price, and Delivery_Days. The columns with the highest number of missing values are Quantity, Price, and Delivery_Days, each having 3 missing entries.
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
```

```python
Explanation

I filled the missing customer names with "Unknown" because the missing names cannot be recovered, but the order records are still useful and should not be removed from the dataset.
```

```python
print(df['Quantity'].describe())
print(df['Price'].describe())
```

```python
quantity_median = df['Quantity'].median()
df['Quantity'] = df['Quantity'].fillna(quantity_median)
```

```python
Explanation

I used the median for Quantity because it is less affected by extreme values and provides a reliable measure of the typical order quantity.
```

```python
#Fill Price with Median
price_median = df['Price'].median()
df['Price'] = df['Price'].fillna(price_median)
```

```python
Explanation

I used the median for Price because the values vary considerably between products, and the median is more robust than the mean when handling potentially skewed data.
```

```python
#Check for Outliers

Q1 = df['Delivery_Days'].quantile(0.25)
Q3 = df['Delivery_Days'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers = df[
    (df['Delivery_Days'] < lower_bound) |
    (df['Delivery_Days'] > upper_bound)
]

print("Outliers:")
print(outliers)
```

```python
#Fill Missing Delivery Days
delivery_median = df['Delivery_Days'].median()
df['Delivery_Days'] = df['Delivery_Days'].fillna(delivery_median)
```

```python
Explanation

The Delivery_Days column does not contain significant outliers based on the IQR method. Since there are no major outliers, the median is an appropriate choice because it represents the central delivery time and is resistant to unusual values.
```

```python
#Handle Missing Country and Product
```

```python
#Country
country_mode = df['Country'].mode()[0]
df['Country'] = df['Country'].fillna(country_mode)
```

```python
Explanation

Country is a categorical column, so filling missing values with the most frequent country (mode) helps preserve the dataset while maintaining realistic values.
```

```python
#Product
product_mode = df['Product'].mode()[0]
df['Product'] = df['Product'].fillna(product_mode)
```

```python
Explanation

Product is also a categorical column. Filling missing values with the most common product allows us to retain all records and avoid losing potentially useful sales information.
```

```python
Reflection Paragraph

The hardest decision was determining how to handle the missing values in the Delivery_Days column because it required checking for outliers before deciding whether the mean or median would be more appropriate. I did not use the same method for every column because different data types require different approaches. For Customer_Name, I used a placeholder value since it is impossible to infer the missing names. For Country and Product, I used the mode because they are categorical variables. For Quantity, Price, and Delivery_Days, I used the median because it is less sensitive to extreme values and provides a stable estimate of the center of the data. If 50% of the Price column were missing, I would be more cautious about simply filling the values with the median because a large amount of missing data can introduce bias. In that situation, I would investigate why the data is missing and consider collecting additional information or using more advanced imputation techniques. This process showed that data cleaning decisions should be based on the characteristics of each column rather than applying a single method to all missing values.
```
