```python
import pandas as pd

data = {
    "Order_ID": ["ORD001", "ORD002", "ORD003", "ORD004", "ORD005",
                 "ORD006", "ORD007", "ORD008", "ORD009", "ORD010"],
    "Product": ["Laptop", "Rice Bag", "Headphones", "Notebook", "Blender",
                "Rice Bag", "Laptop", "Notebook", "Blender", "Headphones"],
    "Category": ["Electronics", "Food", "Electronics", "Stationery", "Appliances",
                 "Food", "Electronics", "Stationery", "Appliances", "Electronics"],
    "Region": ["North", "South", "East", "West", "North",
               "East", "South", "North", "West", "North"],
    "Units_Sold": [5, 30, 12, 50, 8, 20, 3, 80, 6, 15],
    "Unit_Price": [850.00, 12.50, 45.00, 3.20, 65.00,
                   12.50, 850.00, 3.20, 65.00, 45.00],
    "Month": ["January", "January", "February", "February", "March",
              "March", "April", "April", "May", "May"]
}

df = pd.DataFrame(data)
print(df[["Product", "Region"]])
```

```python
print(df[["Order_ID", "Category", "Units_Sold"]])
```

```python
print(df["Unit_Price"])
```

```python
print(df[["Unit_Price"]])
```

```python
print(df[df["Region"] == "North"])
```

```python
print(df[df["Units_Sold"] > 10])
```

```python
print(df[(df["Category"] == "Electronics") & (df["Units_Sold"] > 5)])
```

```python
print(df[(df["Region"] == "South") | (df["Region"] == "West")])
```

```python
print(df.groupby("Region")["Units_Sold"].sum())
```

```python
print(df.groupby("Category")["Unit_Price"].mean())
```

```python
print(
    df.groupby("Category").agg({
        "Units_Sold": "sum",
        "Unit_Price": "mean"
    })
)
```

```python
print(df.groupby("Month")["Units_Sold"].sum())
```

```python
region_totals = df.groupby("Region")["Units_Sold"].sum()

print(region_totals.idxmax())
```

```python
product_prices = df.groupby("Product")["Unit_Price"].mean()

print(product_prices.idxmin())
```

```python
month_totals = df.groupby("Month")["Units_Sold"].sum()

print(month_totals.idxmax())
```

```python
print(
    df[
        (df["Category"] == "Electronics") &
        (df["Region"] == "North")
    ][["Product", "Units_Sold", "Unit_Price"]]
)
```

```python
df["Total_Sales"] = df["Units_Sold"] * df["Unit_Price"]

print(df)
```
